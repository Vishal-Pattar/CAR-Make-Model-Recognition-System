```python
import numpy as np
import pandas as pd
```

```python
from scipy.io import loadmat
from pathlib import Path
```

```python
car_train = Path('./dataset/cars_train')

cars_metadata = loadmat('./car_annotation/cars_meta.mat')
cars_train_annos = loadmat('./car_annotation/cars_train_annos.mat')
```

```python
cars_meta = []
for x in cars_metadata['class_names'][0]:
    cars_meta.append(x)
    
cars_classes = pd.DataFrame(cars_meta,columns=['class_names'])
cars_classes
```

```python
fname = [[x.flatten()[0] for x in i]  for i in cars_train_annos['annotations'][0]]
column_list = ['bbox_x1','bbox_y1','bbox_x2','bbox_y2','class','file_name']
df = pd.DataFrame(fname,columns = column_list)
df['class'] = df['class'] - 1
df['file_name'] = [car_train/i for i in df['file_name']]
df.head()
```

```python
df = df.merge(cars_classes, left_on='class', right_index=True)
df = df.sort_index()
df.sample(10)
```

```python
df = df.reindex(columns=['file_name', 'bbox_x1', 'bbox_y1', 'bbox_x2', 'bbox_y2', 'class', 'class_names'])
df.sample(10)
```

```python
df.to_csv("train_data.csv")
```
