```python
import numpy as np
import pandas as pd
```

```python
from scipy.io import loadmat
from pathlib import Path
```

```python
car_test = Path('./dataset/cars_test')

cars_metadata = loadmat('./car_annotation/cars_meta.mat')
cars_test_annos = loadmat('./car_annotation/cars_test_annos.mat')
```

```python
fname = [[x.flatten()[0] for x in i]  for i in cars_test_annos['annotations'][0]]
column_list = ['bbox_x1','bbox_y1','bbox_x2','bbox_y2','file_name']
df = pd.DataFrame(fname,columns = column_list)
df['file_name'] = [car_test/i for i in df['file_name']]
df.head()
```

```python
df = df.reindex(columns=['file_name', 'bbox_x1', 'bbox_y1', 'bbox_x2', 'bbox_y2'])
df.sample(10)
```

```python
df.to_csv("test_data.csv")
```
